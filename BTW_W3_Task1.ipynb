```python
#imporing the libraries that we might need in this task solution
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#to adjust theme for seaborn plots we will use set theme
sns.set_theme()
```

```python
#importing the titanic dataset from local that was given with the tasks 
file_path = 'titanic.csv'
titanic = pd.read_csv(file_path)
titanic.head(3)
```

```python
#importing another data set of historical prices for practice purpose
#dataset link https://www.wsj.com/market-data/quotes/index/DJIA/historical-prices
file_path = 'HistoricalPrices.csv'
HPrices = pd.read_csv(file_path)
HPrices.head(3)
```

Task # 1.1:
- Create a simple line plot using Matplotlib to display the trend of a list of values
over time.

```python
#lets sort the dataframe based on date
HPrices.sort_values(by='Date',inplace=True)
# Convert the date column to datetime with specified format
HPrices['Date'] = pd.to_datetime(HPrices['Date'],format='%Y-%m-%d')
plt.figure(figsize=(12,4))

plt.title('Close Vs Date Graph')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid(True)

plt.plot(HPrices['Date'],HPrices[' Close'])
plt.show()
```

Task # 1.2:
- Plot a bar chart using Matplotlib to show the frequency of different categories
in a dataset.

```python
grouped_dataset = titanic.groupby(['Pclass','Survived']).size().unstack()
grouped_dataset
```

```python
x_axis = grouped_dataset.index
survived = grouped_dataset[0]
not_survived = grouped_dataset[1]
print(survived)
print(not_survived)
```

```python
fig, (plt1, plt2) = plt.subplots(1, 2, figsize=(10, 4))

plt1.set_title('Survived Vs Passenger Class')
plt1.bar(x_axis,survived,edgecolor = 'Black',color = 'Green')
plt1.set_xlabel('Passenger Class')
plt1.set_ylabel('People Survived')
plt1.set_ylim(0,400)

plt2.set_title('Not Survived Vs Passenger Class')
plt2.bar(x_axis,not_survived,edgecolor = 'Black',color = 'Green')
plt2.set_xlabel('Passenger Class')
plt2.set_ylabel('People Not Survived')
plt2.set_ylim(0,400)
plt.show()
```

Task # 1.6:
- Plot a heatmap using Seaborn to visualize the correlation matrix of a dataset

```python
#for correlation purpose we are gonna import a dataset that is composed of numerical values and to visualize the correlation better
#importing boston house dataset
#i got the dataset with me so i just have to load it here from local
file_path = 'HousingData.csv'
boston = pd.read_csv(file_path)
boston.head()
#now lets create a correlation so that we can visualize it with heatmap so that we can identify positive and negative 
correlation_matrix = boston.corr()
correlation_matrix
```

```python
#now lets visulize it with heatmap
plt.figure(figsize=(12,4))
sns.heatmap(correlation_matrix,annot=True,fmt = '.1f',annot_kws={'size':8},cmap = 'Oranges')
plt.show()
```

Task # 1.7 : The dark values shows better and positive correlation , while the light shows negative correlation 
- In positive correlation both variable increase and decrease together means there is a good direct relationship
- While in negative we have an inverse relation when 1 variable increase the other decreases and vice versa

```python
t_axis = np.linspace(0,5,50)
sin_function = np.sin(np.pi*t_axis)
cos_function = np.cos(np.pi*t_axis)
# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))  # Adjusted figsize for better layout

# Plot in the first subplot
ax1.plot(t_axis, sin_function, 'r+', label='sin function')
ax1.set_xlabel("Time axis")
ax1.set_ylabel("Output function")
ax1.set_title("subplot 1")
ax1.set_xlim(-1, 6)
ax1.set_ylim(-2, 2)
ax1.grid(True)
ax1.legend(loc=(0.6, 0.9))

# Plot in the second subplot
ax2.plot(t_axis, cos_function, 'g+', label='cos function')
ax2.set_xlabel("Time axis")
ax2.set_ylabel("Output function")
ax2.set_title("subplot 2")
ax2.set_xlim(-1, 6)
ax2.set_ylim(-2, 2)
ax2.grid(True)
ax2.legend(loc=(0.6, 0.9))

# Show the plot
plt.tight_layout()
plt.show()
```

Task # 1.8 : Customize the appearance of a Seaborn plot by changing the color palette, 
adding titles, and modifying axis labels.

```python
tips = sns.load_dataset('tips')
#this is a tips dataset from seaborn where we have tips payed against different attributes
plt.figure(figsize=(6,4))
#due to hue it will differentiate automatically with different colours based on the specified attribute
#now style will assign different representation based on specified attribute
#size parameter will differentiate between different categories of specified attribute with the help of different sizes

sns.relplot(x=tips['total_bill'],y=tips['tip'],hue=tips['smoker'],style=tips['smoker'],size=tips['size'])
plt.title('Tip Vs Paid Bill')
plt.xlabel('Bill To be Paid')
plt.ylabel('Tip Given')
plt.show()
```

Task # 1.10:
- Combine Matplotlib and Seaborn to create a complex visualization, such as 
overlaying a KDE plot on a histogram

```python
plt.figure(figsize=(8,4))
plt.title('Close Vs Date Graph')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid(True)

# Plot the line plot
sns.lineplot(x=HPrices['Date'], y=HPrices[' Close'], label='Line Plot')

# Plot the KDE plot
sns.kdeplot(x=HPrices['Date'], y=HPrices[' Close'], label='KDE Plot', cmap='Blues', fill=True)

# Add a legend
plt.legend()

plt.show()
```
